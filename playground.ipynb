```python
from linear_regression import lg_predict
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt(open("./data.csv", "rb"), delimiter=",", skiprows=1)

features = np.delete(data, 3, 1)
y = data[:, 3] + 10  # y = 5*x1 + 3*x2 + x3 + 10 ( c = 10)
y = y.reshape((y.shape[0], 1))

weights, bias, pred, target = lg_predict(features, y, 0.3, 600)
pred = np.squeeze(pred)
target = np.squeeze(target)
error = np.mean(np.abs((pred - target)))
```

```python
plt.plot(pred)
plt.plot(target)
plt.show()
```
